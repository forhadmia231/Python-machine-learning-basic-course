```python
import pandas as pd
import io

# আপনার দেওয়া ডেটা (CSV ফরম্যাটে)
data = """
YEAR,JAN,FEB,MAR,APR,MAY,JUN,JUL,AUG,SEP,OCT,NOV,DEC
1990,3.3,3.17,4.01,5.85,5.27,5.93,6.61,4.92,4.87,4.03,4.42,3.97
1990,3.52,2.88,3.77,5.52,4.95,6.02,6.54,4.89,4.98,4.19,4.36,3.95
... (বাকি ডেটা)
"""

# CSV পড়া
df = pd.read_csv(io.StringIO(data))

# বছর অনুযায়ী প্রতিটি মাসের গড় বের করা
monthly_avg = df.groupby('YEAR').mean().reset_index()

# কলামগুলোর নাম ঠিক করা (JAN, FEB,...)
# এগুলি ইতোমধ্যে ঠিক আছে, শুধু YEAR কলাম থাকবে

# ফলাফল দেখানো
print(monthly_avg.to_string(index=False))

# চাইলে CSV হিসেবে সেভ করতে পারেন
# monthly_avg.to_csv('yearly_monthly_avg_wind.csv', index=False)
```

```python
import pandas as pd

# File paths – adjust if needed
wind_file = "/kaggle/input/datasets/mdforhadmia/monthly-average-weather-dataset/yearly_monthly_avg_wind.csv"
tmin_file = "/kaggle/input/datasets/mdforhadmia/monthly-average-weather-dataset/temperature 2 meter minimum.csv"
tmax_file = "/kaggle/input/datasets/mdforhadmia/monthly-average-weather-dataset/temperature 2 meter maximum.csv"
prec_file = "/kaggle/input/datasets/mdforhadmia/monthly-average-weather-dataset/precipitation average .csv"

# Read the CSV files
df_wind = pd.read_csv(wind_file)
df_tmin = pd.read_csv(tmin_file)
df_tmax = pd.read_csv(tmax_file)
df_prec = pd.read_csv(prec_file)

# Months list (preserves order)
months = ['JAN','FEB','MAR','APR','MAY','JUN','JUL','AUG','SEP','OCT','NOV','DEC']

def melt_and_avg(df, value_name):
    """Melt month columns into rows and compute yearly monthly mean."""
    melted = df.melt(id_vars='YEAR', value_vars=months,
                     var_name='MONTH', value_name=value_name)
    # Average multiple rows (e.g., ensemble members) per year-month
    avg = melted.groupby(['YEAR', 'MONTH'])[value_name].mean().reset_index()
    return avg

# Compute monthly means for each variable
wind_avg   = melt_and_avg(df_wind, 'WIND_SPEED')
tmin_avg   = melt_and_avg(df_tmin, 'TEMP_MIN')
tmax_avg   = melt_and_avg(df_tmax, 'TEMP_MAX')
prec_avg   = melt_and_avg(df_prec, 'PRECIPITATION')

# Merge all on YEAR and MONTH
merged = wind_avg.merge(tmin_avg, on=['YEAR','MONTH'], how='outer') \
                 .merge(tmax_avg, on=['YEAR','MONTH'], how='outer') \
                 .merge(prec_avg, on=['YEAR','MONTH'], how='outer')

# Convert MONTH to ordered categorical so sorting keeps calendar order
merged['MONTH'] = pd.Categorical(merged['MONTH'], categories=months, ordered=True)
merged = merged.sort_values(['YEAR', 'MONTH']).reset_index(drop=True)

# Round numeric columns for readability
merged = merged.round(2)

# Save final dataset
merged.to_csv("monthly_weather_dataset.csv", index=False)

# Show first rows
print(merged.head(12))
```
